### Line Graphs ###

To get us started with line graphs, let's import the `movies_by_year.csv` file which contains data on movies produced by U.S. studios in each of the years 1980 through 2015. 

The columns are:

| **Column** | Content |
|------------|---------|
| `Year` | Year |
| `Total Gross` | Total domestic box office gross, in millions of dollars, of all movies released |
| `Number of Movies` | Number of movies released |
| `#1 Movie` | Highest grossing movie |

````python
#Read in the 'movies_by_year.csv' here.
movies_by_year = pd.read_csv('movies_by_year.csv')
movies_by_year.head(10)
````

Line graphs are among the most common visualizations and are often used to study chronological trends and patterns.

Recall that we already imported the matplotlib plotting library as ```plt```.  We'll rely on ```plt``` to make the visualizations.  The method ```plot``` from the ```plt``` package draws a line graph using one point for each row of the table. 

Sensibly, you must pass these 2 pieces of information to the ``plot`` function:

* first give the column whose data will appear on the horizontal axis 
* and then give the column whose data will appear on the vertical axis. 

#### **IMPORTANT RULE:  For every graph we make in this class, please include a title and axes labels**

Here's how we can plot the number of movies (y) per year (x) and give it a title/labels.

````python
plt.plot(movies_by_year['Year'], movies_by_year['Number of Movies'])  #Input the column for the horizontal data 
                                                                      #and then column for the vertical data
plt.title('Number of Movies Per Year')  #title
plt.xlabel('Year')                      #label x axis 
plt.ylabel('Number of Movies')          #label y axis
````

If you wanted to show a point for every single (x,y) pair in the line graph, you could do so by adding the ```marker="o" ``` parameter below which tells matplotlib that you want a circle at each plotted point.

Be wary of doing this, however.  If it seems to  "clutter" the graph, the don't do it unless you have to, ie, unless your x axis datapoints have unequal intervals.

````python
plt.plot(movies_by_year['Year'], movies_by_year['Number of Movies'], marker='o')
plt.title('Number of Movies Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
````

The graph rises sharply and then has a gentle upwards trend though the numbers vary noticeably from year to year. The sharp rise in the early 1980's is due in part to studios returning to the forefront of movie production after some years of filmmaker-driven movies in the 1970's.  And you know what the sharp decline in 2020 is from....

If desired, you can actually tweak the range of the x and y axis.  For example, even though though line graphs do not have to start at 0, let's say you wanted the y axis to go from 0-1000.  You can use the ``plt.ylim(0,1000)``  function to do this.

Fill in the ellipsis below to change the y axis range as desired.

````python
plt.plot(movies_by_year['Year'], movies_by_year['Number of Movies'])
plt.title('Number of Movies Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')

plt.ylim(0,1000)  #Fill in the ellipses with 0 and 1000 to chang
````

### Noise and Moving/Rolling Averages 

There should have been lots of NOISE in your previous graph, meaning, there are a lot of jagged ups and downs.

Jaggeness isn't necessarily bad, but sometimes we want to smooth out our jaggedness.

One way you could deal with this is to plot at a different interval, say every 3rd day instead of every day.

---

# **Question 8**

Copy and paste your code to question 7 into the cell below.   Then change one line of code according to the following directions.

Change the ``plt.plot(...)`` line of code so that you add ``.iloc[::3]`` to your x/y columns. (So replace the XXX's below with what you graphed.)  

```plt.plot(XXXX.iloc[::3], XXX.iloc[::3])```

Recall  that ``.iloc[::3]`` grads every THIRD row in your data, instead of every row in your data.  

You should notice that the graph looks less "jagged-ey".  Of course, there are still many quick ups and downs so some jaggedness still exists.

````python
#Copy the answer to 7 below.
#Then change the plt.plot line to look like:  plt.plot(XXXX.iloc[::3], XXX.iloc[::3])
plt.figure(figsize =(20,5))
game_stop['Net Change'] = game_stop['Open'] - game_stop['Close']
plt.plot(game_stop['Date'].iloc[: : 3] , game_stop['Net Change'].iloc[: : 3])
plt.xlabel('Date')
plt.ylabel('Net Change')
plt.title('Net Change over Time')
plt.ylim(-100,100)
````

---

Do you agree that the graph looks less "jagged-ey".  Hopefully you do.

On the other hand, graphing only every 3rd row also means that you left out some data, which can feel a bit unnerving. 

So as another way to get rid of noise/jaggedness, you could instead plot a rolling average, meaning, the average over the previous $n$ rows.  Plotting rolling averages  "dampens" the jaggedness, if that is what you are looking to do.

To get started doing this, we can just use the ``rolling`` function, where here we average over the previous 5 rows (well, 4 previous rows plus the current one).  Since we have information for each day in our data, 5 rows corresponds to 5 days and so we're actually calculating 5 day averages in this example:

```game_stop["Net Change"].rolling(5).mean()```

Note:  

 * ``rolling(5)`` accumulates the last 5 rows 
 * and ``.mean()`` takes their average.

If we wanted to plot the rolling average over the last 7 rows/days, we would use:

```game_stop["Net Change"].rolling(7).mean()```
